```python
#检查数据中的重复值
train.duplicated().sum()
```

```python
#去除重复值
train = train.drop_duplicates()
train.duplicated().sum()
```

```python
#去除重复值后，重新恢复索引
train.index = range(train.shape[0])
```

```python
train.info()
```

```python
train.isnull().sum()
```

```python
train.isnull().sum() / train.shape[0]
```

```python
#由于 NumberOfDependents 的缺失值只有2%，可以利用平均值来填充
NOD_mean = train['NumberOfDependents'].mean()
train['NumberOfDependents'].fillna(NOD_mean,inplace=True)
```

```python
train.info()
```

```python
train['MonthlyIncome'].mode()[0]
```

```python
train_show = train
train_show.rename(columns = {'SeriousDlqin2yrs':'未来两年可能违约', 'RevolvingUtilizationOfUnsecuredLines':'可用信贷额度比例', 'age':'年龄',
       'NumberOfTime30-59DaysPastDueNotWorse':'逾期30-59天的笔数', 'DebtRatio':'负债率', 'MonthlyIncome':'月收入',
       'NumberOfOpenCreditLinesAndLoans':'信贷数量', 'NumberOfTimes90DaysLate':'逾期90天+的笔数',
       'NumberRealEstateLoansOrLines':'固定资产贷款数', 'NumberOfTime60-89DaysPastDueNotWorse':'逾期60-89天的笔数',
       'NumberOfDependents':'家属数量'},inplace=True)
```

```python
plt.figure(figsize=(10,5),dpi=300)
sns.heatmap(train_show.corr(),cmap="Reds",annot=True)
```
